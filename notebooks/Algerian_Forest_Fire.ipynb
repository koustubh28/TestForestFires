### Algerian Forest Fires Dataset

#### The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.
#### 122 instances for each region.

#### The period from June 2012 to September 2012.
#### The dataset includes 11 attributes and 1 output attribute (class)
#### The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

### Data Cleaning

```python
# Missing values
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region:

1 : "Bejala Region Dataset"
2 : "Sidi-Bel Abbes Region"

Add new column with region

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## So now we want to remove the unwanted row that starts from 122:
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

We need to change the data types for these columns:

### Change the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

#### Changing other columns to float data type


```python
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
## Let's get the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv', index=False)
```

### Exploratory Data Analysis

```python
# Drop day, month and year
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
## Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
df_copy['Classes'].value_counts()
```

```python
# Density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
percentage = df_copy['Classes'].value_counts(normalize=True) * 100
```

Means approx 56% fire and 43% not fire

```python
## Pie chart plotting
classlabels=["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```

```python
## Monthly Fire Analysis
df_temp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13, 6))
sns.set_style('whitegrid')

# Define your custom color palette
custom_colors = {'fire': '#E63946', 'not fire': 'lightgreen'}

# Pass the palette to the countplot
sns.countplot(x='month', hue='Classes', data=df, palette=custom_colors)
plt.title("Fire Analysis of bejai Region")
```

#### It's observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things.
